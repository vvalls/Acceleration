```julia
k = 2;
upper_bound = zeros(k)
lower_bound = zeros(k)
s = zeros(k)
A = 0;

p = -0.5;

for i=1:k
   α = i^p;
   A = A + α;
   s[i] = A;
   upper_bound[i] = (i+1)^(p+1) / (p+1)
   lower_bound[i] = (i)^(p+1) / (p+1) - 1/(p+1)
end
```

```julia
using Plots
plot(hcat(upper_bound,lower_bound,s))
```
