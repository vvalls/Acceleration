```julia
A = 1;
B = A;
L = 1;
μ = 0;
σ = 1;

k = 100;
log_A = zeros(k)
log_B = zeros(k)

for i=1:k
    a = L; b = -(μ*A + σ); c = -μ*A^2 - σ*A;
    α = (-b + sqrt(b^2 - 4*a*c)) / (2*a);
    A = A + α;
    
    β = σ/(2L) * (1 + 2 * sqrt(L/σ) * sqrt(B)) 
    β = σ/(2L) * (1 + i) 
    β = sqrt( i*(i-2)/2 + 1)
    B = B + β;
    #println(B)
    log_A[i] = A;
    log_B[i] = B;
end

using Plots
plot(hcat(log_A,log_B))#,yaxis=:log)




```
