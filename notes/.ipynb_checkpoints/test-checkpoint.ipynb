```julia
using Acceleration
using LinearAlgebra

n = 1000; # number of samples
d = 50;    # number of features
selected_labels = [1,2];  ## Only two labels allowed

data, labels = load_dataset("MNIST",[0,1],n,d);
f,∇f,∇ϕ_cjg,σ,L,μ  = linear_regression(data,labels);
μ2 = μ
```

```julia
n = size(data,2)
x_ini = zeros(n)


using JuMP
using Ipopt

@time model = Model(with_optimizer(Ipopt.Optimizer))
@time set_silent(model)
@time @variable(model, x[1:n])
@time @objective(model, Min, f(x))
@time optimize!(model)
@time x_opt = value.(x);
```

```julia
n = length(x_ini)
x = x_ini;
y = zeros(n);
z = zeros(n);
v = zeros(n);
A = 0;
AX = zeros(n);

k = 10000;
yks = zeros(n,k);
zks = zeros(n,k);
var = zeros(k)
noise = zeros(k);
noise2 = zeros(k);
μ = 0;
nn = 0;
nn2 = 0;#norm(x_opt)^2;
AN = zeros(n);
AF = zeros(n);

an = zeros(k);
af = zeros(k);
anf = zeros(k);

for i=1:k   
    a = (L-μ); b = -(2*μ*A+σ); c = -μ*A^2 - σ*A;
    α = (-b + sqrt(b^2 - 4*a*c)) / (2*a);
    A = A + α;
    θ = α * inv(A);
    x = inv(A/α - μ*α/(μ*A+σ)) * ((A-α)/α *y + (μ*(A-α)+σ)/(μ*A + σ) * v);
    AX = AX + α*x;

    #η = -∇f(x) + sign.(∇f(x)) .* (abs.(∇f(x)) .> norm(∇f(x),Inf)/2) .* norm(∇f(x),Inf)/2;
    η = (1/(3*α)) * (z - AN - α*∇f(x));
    
    #η = inv(3*α)*AF - α*∇f(x)
    
    AF = AF + α*∇f(x);
    
    #η = (1/(2*α)) * (z - x_opt - α*∇f(x))
    #η = zeros(n);
    
    AN = AN + α*η;

    approx_gradient = ∇f(x) + η
    z = z - α*approx_gradient;
    #println(norm(η))
    
    v = (z + μ*AX)*inv(1 + μ*A);
    y = (A-α)*inv(A)*y + α*inv(A)*v; 

    var[i] = norm(x_opt-z)^2;#α*η'*(x_opt[:,1] - v);
    yks[:,i] = y;
    zks[:,i] = z;
    
    nn = nn + α*η'*(x_opt-z);
    noise[i] = nn;
    anf[i] = norm(AF + 2*AN);
    nn2 = nn2 + α*η'*(AN - z) - 0.5*α^2 * η'*η;
    noise2[i] = nn2;

end

x_out3 = yks;
x_out2 = gradient_descent(f,∇f,∇ϕ_cjg,x_ini,σ,L,k);
x_out4 = accelerated_mirror_descent(f,∇f,∇ϕ_cjg,x_ini,σ,L,k);
x_out5 = SCVX_AMD(f,∇f,∇ϕ_cjg,x_ini,σ,L,μ2,k);
```

```julia
using Plots
using LinearAlgebra

p2 = sqrt.(sum((x_out2 .- x_opt).^2, dims=1)');
p3 = sqrt.(sum((x_out3 .- x_opt).^2, dims=1)');
p4 = sqrt.(sum((x_out4 .- x_opt).^2, dims=1)');
p5 = sqrt.(sum((x_out5 .- x_opt).^2, dims=1)');
pz = sqrt.(sum((zks .- x_opt).^2, dims=1)');
#pz = sqrt.(sum((zks .- x_opt).^2, dims=1)');
#plot(hcat(p3,p4),yaxis=:log)
plot(max.(hcat(p2,p4,p5,p3),1e-20),yaxis=:log,linewidth=2,xlabel="training rounds",ylabel="loss")
```
