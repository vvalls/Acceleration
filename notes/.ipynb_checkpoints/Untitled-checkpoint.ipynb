```julia
A = 1;
B = 1;
L = 10;
μ = 0;
σ = 1;

k = 1000;
log_A = zeros(k)
log_B = zeros(k)

for i=1:k
    a = L; b = -(μ*A + σ); c = -μ*A^2 - σ*A;
    α = (-b + sqrt(b^2 - 4*a*c)) / (2*a);
    A = A + α;
    
    #β = (1/2)*(σ/L) * (1+sqrt((σ/(2*L))^2 + σ/L * B))
    β = σ/(2L) * (1 + 2 * sqrt(L/σ) * sqrt(B))  #0.5*(σ/L) + sqrt(σ^2 / (4*L) + σ/L * B)
    B = B + β;
    
    log_A[i] = A;
    log_B[i] = B;
end

using Plots
plot(hcat(log_A,log_B))#,yaxis=:log)




```
