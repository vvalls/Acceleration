```julia
using Acceleration
using LinearAlgebra

data, labels = load_libsvm_dataset("mushrooms",500); # dataset name (a5a, mushrooms), num samples
f,∇f,∇ϕ_cjg,σ,L,μ  = logistic_regression(data,labels,1);
```

```julia
d,n = size(data)
x_ini = zeros(n)
k = 1000;
μ = 1
L = 500
ω = 0; 

A = data;
b = labels;
m = 2;
θ = 1

function exact_gradient(x)
    return ∇f(x), 0
end

function dist_f(x)
    
    sample_size = Int16(floor(size(A,1)/m)) 
    num_features = size(A,2)
    
    sumf = 0;
    
    for j=1:m
        index1 = (j-1)*sample_size + 1;
        index2 = j*sample_size;
        Aj = A[index1:index2,:];
        bj = b[index1:index2];
        sumf = sumf + sum(log.(1 .+ exp.(-bj.*(Aj*x))))
    end
    
    return sumf + (θ/2)*norm(x,2)^2
end


function distributed_gradient(x)

    sample_size = Int16(floor(size(A,1)/m)) 
    num_features = size(A,2)

    gradients = zeros(m,num_features)

    for j=1:m
        index1 = (j-1)*sample_size + 1;
        index2 = j*sample_size;
        Aj = A[index1:index2,:];
        bj = b[index1:index2];
        grad  = ((exp.(-bj.*Aj*x) ./ (1 .+ exp.(-bj.*Aj*x)) )'*(-bj.*Aj))'[:,1] + (1/m)*θ*x
        gradients[j,:] = grad
    end 
    
    return sum(gradients,dims=1)'[:,1],0;
end

function distributed_gradient_natural(x)
    
    sample_size = Int16(floor(size(A,1)/m)) 
    num_features = size(A,2)

    gradients = zeros(m,num_features)

    for j=1:m
        index1 = 1; #(j-1)*sample_size + 1;
        index2 = sample_size*m; #j*sample_size;
        Aj = A[index1:index2,:];
        bj = b[index1:index2];
        grad  = ((exp.(-bj.*Aj*x) ./ (1 .+ exp.(-bj.*Aj*x)) )'*(-bj.*Aj))'[:,1] + θ*x
        gm = size(grad,1)
        grad = grad  + 0.0001*randn(gm)
        nat_grad = sign.(grad).*2.0.^(floor.(log2.(abs.(grad)))).*(1 .+ rand(gm))
        gradients[j,:] = nat_grad 
    end 
    
    return sum(gradients,dims=1)'[:,1] , 0;
end


function natural_compression(x)

    grad = ∇f(x)
    n = size(grad,1)
    nat_grad = sign.(grad).*2.0.^(floor.(log2.(abs.(grad)))).*(1 .+ rand(n))
    
    return nat_grad, norm(nat_grad,2)
end

grad_fun = distributed_gradient_natural
fun = dist_f

@time f_opt = ufom(fun,distributed_gradient,∇ϕ_cjg,x_ini,σ,L,μ,2*k,Inf)[end]

@time p1 = MD(fun,grad_fun,∇ϕ_cjg,x_ini,σ,L,k) .- f_opt;
@time p2 = PGD_avg(fun,grad_fun,∇ϕ_cjg,x_ini,σ,L,k) .- f_opt;
@time p3 = ufom(fun,grad_fun,∇ϕ_cjg,x_ini,σ,L,μ,k,Inf) .- f_opt; # Our algorithm
@time p4 = AMD_plus(fun,grad_fun,∇ϕ_cjg,x_ini,σ,L,k) .- f_opt;
@time p5 = μAMD_plus(fun,grad_fun,∇ϕ_cjg,x_ini,L,μ,k) .- f_opt;
#@time p6 = ACGD(f,grad_fun,x_ini,L,k,ω) .- f_opt;
@time p7 = μACGD(fun,grad_fun,x_ini,L,μ,k,ω) .- f_opt;
#@time p8 = Nesterov83(fun,grad_fun,x_ini,k) .- f_opt;
```

```julia
using Plots

fig = plot(max.(hcat(p1,p2,p3,p4,p5,p7),1e-16),yaxis=:log, linewidth=2, legend=:topright,
label = ["Mirror-descent" "Gradient-descent (avg)" "New Algorithm" "AMD+ (ICML 2018)" "μAMD+ (ICML 2018)" "ACGD (ICML 2020)" "AGD (Nesterov)"], xlabel = "Training rounds", ylabel = "Loss")
```

```julia
m = 10

function natural_compression(x)
    
    sample_size = Int16(floor(size(A,1)/m)) 
    num_features = size(A,2)

    gradients = zeros(m,num_features)

    for j=1:m
        index1 = 1; #(j-1)*sample_size + 1;
        index2 = sample_size*m; #j*sample_size;
        Aj = A[index1:index2,:];
        bj = b[index1:index2];
        grad  = ((exp.(-bj.*Aj*x) ./ (1 .+ exp.(-bj.*Aj*x)) )'*(-bj.*Aj))'[:,1] + θ*x
        gm = size(grad,1)
        grad = (1/m)*grad  #+ 0.0001*randn(gm)
        nat_grad = sign.(grad).*2.0.^(floor.(log2.(abs.(grad)))).*(1 .+ rand(gm))
        gradients[j,:] = nat_grad 
    end 
    
    return sum(gradients,dims=1)'[:,1] , 0;
end

function random_sparsification(x)
    
    sample_size = Int16(floor(size(A,1)/m)) 
    num_features = size(A,2)

    gradients = zeros(m,num_features)

    for j=1:m
        index1 = 1; #(j-1)*sample_size + 1;
        index2 = sample_size*m; #j*sample_size;
        Aj = A[index1:index2,:];
        bj = b[index1:index2];
        grad  = ((exp.(-bj.*Aj*x) ./ (1 .+ exp.(-bj.*Aj*x)) )'*(-bj.*Aj))'[:,1] + θ*x
        gm = size(grad,1)
        grad = (1/m)*grad  #+ 0.0001*randn(gm)
        gm = size(grad,1)
        r = 1/4
        mask = Float64.(rand(gm) .< r)
        rs_gradient = (1/r)*grad.*mask
        gradients[j,:] = rs_gradient 
    end 
    
    return sum(gradients,dims=1)'[:,1] , 0;
end


function dithering(x)
    
    sample_size = Int16(floor(size(A,1)/m)) 
    num_features = size(A,2)

    gradients = zeros(m,num_features)

    for j=1:m
        index1 = 1; #(j-1)*sample_size + 1;
        index2 = sample_size*m; #j*sample_size;
        Aj = A[index1:index2,:];
        bj = b[index1:index2];
        grad  = ((exp.(-bj.*Aj*x) ./ (1 .+ exp.(-bj.*Aj*x)) )'*(-bj.*Aj))'[:,1] + θ*x
        gm = size(grad,1)
        grad = (1/m)*grad  #+ 0.0001*randn(gm)
        q = Inf;
        gm = size(grad,1)
        levels = 10;
        leveled_gradient = sign.(grad).*(levels/norm(grad,q)).*grad/2
        base = floor.(leveled_gradient)
        rv = Float64.(rand(gm) .> (leveled_gradient .- floor.(leveled_gradient)))
        quantized_gradient = sign.(grad).*(base .+ rv)*norm(grad,q)/levels*2;
        gradients[j,:] = quantized_gradient 
    end 
    
    return sum(gradients,dims=1)'[:,1] , 0;
end


function dist_f(x)
    
    sample_size = Int16(floor(size(A,1)/m)) 
    num_features = size(A,2)
    
    sumf = 0;
    
    for j=1:m
        index1 = (j-1)*sample_size + 1;
        index2 = j*sample_size;
        Aj = A[index1:index2,:];
        bj = b[index1:index2];
        sumf = sumf + sum(log.(1 .+ exp.(-bj.*(Aj*x))))
    end
    
    return sumf + (θ/2)*norm(x,2)^2
end
```

```julia
using Acceleration
using LinearAlgebra

for model in ["mushrooms"]
    
    data, labels = load_libsvm_dataset(model,1000); # dataset name (a5a, mushrooms), num samples
    f,∇f,∇ϕ_cjg,σ,L,μ  = logistic_regression(data,labels,1);

    d,n = size(data)
    x_ini = zeros(n)
    k = 2000;
    μ = 1
    L = 200
    p3 = zeros(k)
    p4 = zeros(k)
    p7 = zeros(k)
    p9 = zeros(k)
    
    fun = dist_f

    @time f_opt = ufom(fun,exact_gradient,∇ϕ_cjg,x_ini,σ,L,μ,2500,Inf)[end]

    @time for grad_fun in [exact_gradient,random_sparsification, dithering, natural_compression]

        @time p3 = max.(1e-16,ufom(fun,grad_fun,∇ϕ_cjg,x_ini,σ,L,μ,k,Inf) .- f_opt);
        p4 = max.(1e-16,AMD_plus(fun,grad_fun,∇ϕ_cjg,x_ini,σ,L,k) .- f_opt);
        p7 = max.(1e-16,μACGD(fun,grad_fun,x_ini,L,μ,k,0) .- f_opt);
        p9 = max.(1e-16,ufom(fun,grad_fun,∇ϕ_cjg,x_ini,σ,L,μ,k,5) .- f_opt);
    
        filename = string("logs/",model,"_",grad_fun,".txt")
        io = open(filename, "w")
        
        if(grad_fun==exact_gradient)
            bits = n*64;
        elseif(grad_fun==random_sparsification)
            bits = (n/4)*64;
        elseif(grad_fun==dithering)
            bits = (2.8*n + 64);
        elseif(grad_fun==natural_compression)
            bits = n*9;
        end
        
        for i=1:k
            print(io,i)
            print(io,"\t")
            print(io,i*n*m*64)
            print(io,"\t")
            print(io,i*m*bits)
            print(io,"\t")
            print(io, p7[i])
            print(io,"\t")
            print(io, p4[i])
            print(io,"\t")
            print(io, p3[i])
            print(io,"\t")
            print(io, p9[i])
            print(io,"\n")
        end
        close(io)
    end
end
```
